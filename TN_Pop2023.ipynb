```python
import numpy as np
import pandas as pd
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
import seaborn as sns



%matplotlib inline
```

```python
pop = pd.read_csv('tn_pop.csv')
```

```python
pop.head()
```

```python
# Extract the county names and population data for 2010 and 2023
counties = pop['name']
pop2010 = pop['pop2010'].head(10)
pop2023 = pop['pop2023'].head(10)

# Calculate the percent change in population from 2010 to 2023
percent_change = (pop2023 - pop2010) / pop2010 * 100

# Create a scatter plot showing the change in population for each county
plt.scatter(pop2010, pop2023)
for i, county in enumerate(counties):
    plt.annotate(county, (pop2010[i], pop2023[i]))
plt.title('Population Change from 2010 to 2023 for TN Counties')
plt.xlabel('Population in 2010')
plt.ylabel('Population in 2023')
plt.show()
```
